```python
import json
import pymongo
import unittest
```

```python
# load the data of company_profile.json file in a list

with open('company_profiles.json') as json_file:
    company_profiles = json.load(json_file)

company_profiles[:2]
```

```python
company_without_location_info = []
company_without_revenue_info = []
company_without_employee_size_info = []
company_without_website_info = []
company_without_contact_details_info = []

missing_item_list = []
searching_items = ['revenue', 'employee size', 'location', 'website', 'contact details']

for company in company_profiles:
    company_revenue = company.get('company_revenue', None)
    company_employee_size = company.get('company_employee_size', None)
    company_location = company.get('company_location', None)
    company_website = company.get('company_website', None)
    contact_details = company.get('contact_details', None)
    
    if not company_revenue:
        company_without_revenue_info.append(company['company_name'])
        
    if not company_employee_size:
        company_without_employee_size_info.append(company['company_name'])
        
    if not company_location:
        company_without_location_info.append(company['company_name'])
        
    if not company_website:
        company_without_website_info.append(company['company_name'])
        
    if not contact_details:
        company_without_contact_details_info.append(company['company_name'])
        
missing_item_list = [company_without_revenue_info, company_without_employee_size_info,
                        company_without_location_info, company_without_website_info, 
                        company_without_contact_details_info]

for index, item in enumerate(searching_items):
    print("\nNumber of companies which do not have any '{}' information: {}"
          .format(item, len(missing_item_list[index])))
    if missing_item_list[index]:
        print("Some of them are:\n", missing_item_list[index][:5])
```

```python
#Establishing a Connection

myclient = pymongo.MongoClient("mongodb://localhost:27017/")
database = myclient["leadbook"]

collection = database["company_profiles"]

company = collection.find_one({'company_name': 'A + Communications and Security'})
company
```

```python
class TestCompanyInfo(unittest.TestCase):

    def test_company_name(self):
        self.assertEqual('A + Communications and Security', company['company_name'])
    
    def test_company_website(self):
        self.assertEqual('http://www.onlineaplus.com', company['company_website'])
    
    def test_company_revenue(self):
        self.assertEqual('$0 - 1M', company['company_revenue'])
    
    def test_company_employee_size(self):
        self.assertEqual('25 - 100', company['company_employee_size'])
    
    def test_company_industry(self):
        self.assertEqual('Telecommunications', company['company_industry'])
    
    def test_company_webdomain(self):
        self.assertEqual('onlineaplus.com', company['company_webdomain'])


if __name__ == '__main__':
    unittest.main(argv=['ignored', '-v'], exit=False)
```
