```python
# load the data of company_profile.json file in a list

with open('company_profiles.json') as json_file:
    company_profiles = json.load(json_file)

company_profiles[:2]
```

```python
profile_schema = {
    "company_name": {"type": "string", "required": True},
    "company_revenue": {"type": "string"},
    "company_employee_size": {"type": "string"},
    "company_industry": {"type": "string"},
    "company_location": {"type": "string"},
    "company_website": {"type": "string"},
    "company_webdomain": {"type": "string"},
    "contact_details": {"type": "list", "default": None, 'required': False}
}
validator = Validator()

# check if all required information are stored in a correct format for company_profiles.json
for company in company_profiles:
    success = validator.validate(company, profile_schema)

    if not success:
        print("company name: ", company['company_name'])
        print("Errors: ", validator.errors) 
```

```python
contact_schema = {
    "contact_name": {"type": "string", "required": True},
    "contact_jobtitle": {"type": "string", "required": True},
    "contact_email_domain": {"type": "string", "required": True},
    "contact_profile_link": {"type": "string", "required": True},
    "contact_department": {"type": "string"},
    "contact_location": {"type": "string"}
}
validator = Validator()

# check if all required information for contact details are stored in a correct format in company_profiles.json
for company in company_profiles:
    if company['contact_details']:
        for contact in company['contact_details']:
            success = validator.validate(contact, contact_schema)

            if not success:
                print("company name: ", company['company_name'])
                print("Errors: ", validator.errors) 
```
